```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0,100,1000)

m_1 = 1*t
m_2 = 4 *(t-30)

fix, ax = plt.subplots()
plt.title("M1 Generates 1kj/day and M2 4kj/day start date for m1:april 1st and m2 :may 1st")
plt.xlabel("no of day")
plt.ylabel('kj produced per day')
ax.set_xlim(0,50)
ax.set_ylim(0,90)
ax.plot(t, m_1, c='green')
ax.plot(t, m_2, c='brown')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```
